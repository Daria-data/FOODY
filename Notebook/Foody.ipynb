```python
# définir la fct sql_to_pandas

def sql_to_df(sql_query):

    # Use pandas to pass sql query using connection form SQLite3
    df = pd.read_sql(sql_query, conn)

    # Show the resulting DataFrame
    return df
```

```python
#1.Lister les employés (nom et prénom) étant "Representative" ou étant basé au Royaume-Uni (UK)
query = '''SELECT nom, prenom
FROM employe
WHERE fonction = "Representative"
UNION
SELECT nom, prenom
FROM employe
WHERE pays = "UK"
ORDER BY 1'''
sql_to_df(query)
```

```python
#2.Lister les clients (société et pays) ayant commandés via un employé situé à Londres ("London" pour rappel) ou ayant été 
#livré par "Speedy Express"
query = '''SELECT societe, pays
FROM clients
WHERE codeCli IN 
(SELECT codeCli FROM commande WHERE noEmp IN
(SELECT noEmp FROM employe WHERE ville = "London" ))
UNION
SELECT societe, pays
FROM clients
WHERE codeCli IN 
(SELECT codeCli FROM commande WHERE noMess IN
(SELECT noMess FROM messager WHERE nomMess = "Speedy Express" ))
ORDER BY 1'''
sql_to_df(query)
```
